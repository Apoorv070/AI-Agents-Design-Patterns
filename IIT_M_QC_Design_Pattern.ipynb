# Architecting Cognitive Enterprise Systems: A Comprehensive Technical Analysis of Agentic Design Patterns in Human Resources Automation

## 1. Introduction: The Paradigm Shift to Agentic Workflows



<p>
The trajectory of artificial intelligence has shifted dramatically from static,
prompt-response interactions to dynamic, goal-oriented behaviors. This evolution
marks the transition from <strong>Generative AI</strong>—capable of producing text
or code on demand—to <strong>Agentic AI</strong>, systems defined by their autonomy,
reasoning capabilities, and ability to execute complex, multi-step workflows with
minimal human intervention.
</p>

<p>
In the enterprise context, this shift allows organizations to move beyond simple
automation (such as Robotic Process Automation or RPA) toward
<strong>cognitive architectures</strong> that can plan, reflect, collaborate, and
adapt to changing variables.
</p>

<p>
The implementation of agentic systems requires a rigorous understanding of
<strong>architectural design patterns</strong>. Just as Object-Oriented Programming
(OOP) relies on patterns like <em>Singleton</em> or <em>Factory</em>, Agentic AI relies
on structural configurations that dictate how Large Language Models (LLMs)
interact with tools, memory, and one another.
</p>

<p>
These patterns—ranging from atomic single-agent loops to emergent swarm
behaviors—directly influence a system’s <strong>reliability</strong>,
<strong>latency</strong>, and <strong>cost-effectiveness</strong>.
</p>

<p>
This report provides an exhaustive technical analysis of thirteen distinct agentic
patterns:
<strong>Single-Agent</strong>, <strong>Multi-Agent</strong>, <strong>Sequential</strong>,
<strong>Parallel</strong>, <strong>Loop</strong>, <strong>Iterative Refinement</strong>,
<strong>Review and Critique</strong>, <strong>Human-in-the-Loop</strong>,
<strong>Coordinator</strong>, <strong>Hierarchical Task Decomposition</strong>,
<strong>ReAct</strong>, <strong>Swarm</strong>, and <strong>Custom Logic</strong>.
</p>

<p>
These patterns are explored through a unified, high-stakes enterprise use case:
the Human Resources (HR) recruitment pipeline, hereafter referred to as
<strong>“TalentFlow.”</strong>
</p>


### 1.1 Use Case Selection: HR Recruitment vs. Ticket Management

<p>
To demonstrate the full spectrum of agentic capabilities, two potential domains
were evaluated: <strong>Ticket Management (IT Support)</strong> and
<strong>HR Recruitment</strong>.
</p>

<p>
<strong>Ticket Management</strong> is characterized by high-volume, highly structured
data. While it benefits from agentic patterns such as
<strong>Coordinator</strong> (for classification and routing) and
<strong>ReAct</strong> (for resolution and tool-based troubleshooting),
the underlying workflows are typically linear and deterministic.
</p>

<p>
In most ticketing systems, complexity arises primarily from
<strong>database retrieval</strong> and rule-based decision trees rather than
deep reasoning, subjective evaluation, or multi-stakeholder coordination.
</p>

<p>
<strong>HR Recruitment</strong>, by contrast, represents a
<strong>complex adaptive system</strong>. It involves multiple dimensions of
uncertainty, judgment, and interaction, including:
</p>

<ul>
  <li>
    <strong>Unstructured Data Processing:</strong>
    Parsing and interpreting diverse resume formats.
  </li>
  <li>
    <strong>Subjective Reasoning:</strong>
    Evaluating attributes such as cultural fit and leadership potential.
  </li>
  <li>
    <strong>Multi-Modal Sourcing:</strong>
    Interacting with heterogeneous external platforms such as LinkedIn and GitHub.
  </li>
  <li>
    <strong>High-Stakes Decision Making:</strong>
    Making choices that directly impact financial investment and human capital.
  </li>
  <li>
    <strong>Complex Human Interaction:</strong>
    Managing scheduling, negotiation, and multi-level approvals.
  </li>
</ul>

<p>
The HR domain therefore requires not merely automation, but
<strong>augmentation of human judgment</strong>. It must handle ambiguity,
conflicting signals, and evolving constraints—making it an ideal environment
for demonstrating advanced agentic patterns.
</p>

<p>
In particular, HR recruitment naturally exposes the need for patterns such as
<strong>Swarm</strong> (to support hiring committee deliberation),
<strong>Review and Critique</strong> (to mitigate bias and ensure fairness),
and <strong>Human-in-the-Loop</strong> (for final offer authorization and accountability).
</p>

<p>
Consequently, this report focuses exclusively on the
<strong>“TalentFlow”</strong> recruitment architecture as the primary vehicle
for illustrating the full range of agentic design patterns.
</p>

## 2. Foundational Pattern


<p>
Foundational patterns represent the <strong>atomic units of agentic architecture</strong>.
They establish the baseline interaction between the <strong>Large Language Model (LLM)</strong>,
the user prompt, and structured data.
</p>



### 2.1 Single-Agent Pattern

![QuickCURD](https://docs.cloud.google.com/static/architecture/images/choose-design-pattern-agentic-ai-system-single-agent.svg)<br>
Credits: Google



<p>
<strong>Definition:</strong>
The <strong>Single-Agent</strong> pattern is the simplest form of agentic behavior.
It involves a solitary LLM invocation or a strictly linear chain in which the model
processes input and produces output.
</p>

<p>
While this pattern lacks internal feedback loops, memory, or peer collaboration,
it is highly effective for <strong>deterministic tasks</strong> where
<strong>speed</strong>, <strong>predictability</strong>, and
<strong>reliability</strong> are paramount.
</p>

<p>
<strong>Role in the TalentFlow Pipeline: Resume Ingestion</strong><br/>
The entry point of any hiring system is the ingestion of unstructured resume data.
Candidates submit resumes in PDF, Word, or plain-text formats. The system must
extract specific fields—such as <strong>Name</strong>, <strong>Email</strong>,
<strong>Skills</strong>, and <strong>Experience</strong>—to populate the
Applicant Tracking System (ATS).
</p>

<p>
A <strong>Single-Agent with Structured Output</strong> is ideal in this context
because the task is <em>extraction</em>, not <em>reasoning</em>. At this stage,
the agent is not expected to evaluate, rank, or judge the candidate; it simply
transforms unstructured content into a reliable, machine-readable representation.
</p>

<p>
<strong>Technical Implementation:</strong><br/>
This pattern is implemented using <strong>LangChain’s</strong>
<code>with_structured_output</code> method. This approach leverages the native
function-calling capabilities of modern LLMs to constrain the model’s response
to a strict <strong>JSON schema</strong> defined via a
<strong>Pydantic model</strong>.
</p>

<p>
By enforcing schema-level guarantees, this technique minimizes hallucination risk
and ensures downstream systems receive consistently structured data.
</p>


```python
from langchain_core.prompts import ChatPromptTemplate
from pydantic import BaseModel, Field
from typing import List

# --- Step 1: Define the Data Schema ---
# Pydantic allows us to enforce strict data typing.
# If the LLM fails to generate a valid list or integer, the parser will raise an error.
class ResumeData(BaseModel):
    full_name: str = Field(description="The full name of the candidate.")
    email: str = Field(description="The candidate's email address.")
    core_skills: List[str] = Field(description="A list of technical skills mentioned.")
    years_experience: int = Field(description="Total years of professional experience inferred from the text.")
    summary: str = Field(description="A concise 2-sentence professional summary.")

# --- Step 2: Construct the Prompt ---
# The prompt is directive, focusing solely on extraction without interpretation.
extraction_prompt = ChatPromptTemplate.from_template(
    """
    You are an expert Resume Parser for an enterprise HR system.
    Analyze the following resume text and extract the required fields.
    Do not hallucinate information not present in the text.

    Resume Text:
    {resume_text}
    """
)

# --- Step 3: Bind the Model to the Schema ---
# This configures the LLM to output valid JSON matching the ResumeData class.
structured_parser = llm.with_structured_output(ResumeData)

# --- Step 4: Create the Execution Chain ---
parser_chain = extraction_prompt | structured_parser

# --- Tutorial Execution ---
sample_resume = """
ALICE ENGINEER
alice.eng@example.com | (555) 010-9988
Summary: Senior Backend Developer with a focus on scalable distributed systems.
Experience:
- Senior Dev at CloudCorp (2018-Present): Led migration to Kubernetes. Expert in Python, Go, and AWS.
- Junior Dev at StartUp Inc (2015-2018): Java development.
"""

# Invoke the agent
parsed_profile = parser_chain.invoke({"resume_text": sample_resume})
print(parsed_profile)
# Display Results
print(f"--- Parsed Profile ---")
print(f"Name: {parsed_profile.full_name}")
print(f"Email: {parsed_profile.email}")
print(f"Skills: {parsed_profile.core_skills}")
print(f"Experience: {parsed_profile.years_experience} years")
```

<p>
<strong>Architectural Insight:</strong>
The <strong>Single-Agent</strong> pattern enables
<strong>interface standardization</strong> early in the pipeline. By converting
raw, unstructured text into a <strong>Pydantic object</strong> at the point of
ingestion, downstream agents can operate on
<strong>sanitized, schema-validated data</strong> rather than repeatedly
grappling with natural language parsing.
</p>

<p>
This design significantly reduces the <strong>cognitive load</strong> placed on
subsequent, more expensive agentic patterns—allowing them to focus on
reasoning, coordination, and decision-making instead of data cleanup.
</p>


### 2.2 ReAct Pattern (Reasoning + Acting)

![QuickCURD](https://docs.cloud.google.com/static/architecture/images/choose-design-pattern-agentic-ai-system-react.svg)<br>
Credits: Google

<p>
<strong>Definition:</strong>
Introduced by Yao et al. (2022), the <strong>ReAct</strong> pattern functions as a
cognitive bridge that synergizes <strong>Reasoning</strong> (Thought) and
<strong>Acting</strong> (Tool Execution). Unlike a standard linear chain, a ReAct
agent operates in an iterative loop: it <strong>observes</strong> the current state,
<strong>reasons</strong> about what is missing, <strong>selects a tool</strong> to
acquire that information, <strong>observes</strong> the tool output, and then
<strong>updates</strong> its reasoning based on new evidence.
</p>

<p>
This design enables the agent to solve <strong>dynamic problems</strong> where the
solution path is not known a priori and must be discovered through tool-mediated
interaction with external systems.
</p>

<p>
<strong>Role in the TalentFlow Pipeline: Candidate Verification</strong><br/>
Once a candidate profile has been parsed into structured form, high-value claims
must be verified. For example, if a candidate claims
<strong>“Core Contributor to LangChain”</strong>, a static LLM cannot reliably confirm
this because model training data has a cutoff and may be outdated.
</p>

<p>
A <strong>ReAct agent</strong> can perform a lightweight “background check” by
reasoning along the following path:
<strong>“I need evidence of GitHub contributions → I will use a Search Tool → I will
analyze the results → I will confirm or deny the claim.”</strong>
</p>

<p>
<strong>Technical Implementation:</strong><br/>
This pattern is implemented using
<code>langgraph.prebuilt.create_react_agent</code>, which simplifies graph
construction. The agent maintains a <strong>message state</strong>, appending
tool calls and tool outputs to the conversation history across iterations until it
produces a <strong>final verification result</strong>.
</p>


![https___substack-post-media.s3.amazonaws.com_public_images_d02b2eaa-16c3-4f92-8f97-06329fbcccd4_716x550[1].gif]()

```python
from langchain_community.tools import DuckDuckGoSearchRun
from langgraph.prebuilt import create_react_agent
from langchain_core.messages import HumanMessage

# --- Step 1: Define Tools ---
# The agent needs external capabilities. Here, we give it a search engine.
search_tool = DuckDuckGoSearchRun()
tools = [search_tool]

# --- Step 2: Initialize the ReAct Agent ---
# The create_react_agent function automatically constructs the graph with:
# - A model node (to reason and call tools)
# - A tool node (to execute the tools)
# - A conditional edge (loop back if tool called, end if final answer found)
background_check_agent = create_react_agent(llm, tools=tools)

# --- Tutorial Execution ---
# We enable streaming to visualize the "Thought -> Action -> Observation" loop.
query = "Verify if 'Harrison Chase' is associated with the LangChain framework using web search."

print(f"--- Starting ReAct Verification Process ---")
inputs = {"messages": [HumanMessage(content=query)]}

# Stream the steps to show the agent's reasoning process
for step in background_check_agent.stream(inputs, stream_mode="values"):
    message = step["messages"][-1]
    if hasattr(message, 'tool_calls') and message.tool_calls:
        # Access the first ToolCall object and its attributes
        print(f"\n: Need to use tool: {message.tool_calls[0]['name']}")
        print(f": {message.tool_calls[0]['args']}")
    elif message.type == "tool":
        print(f"\n: {message.content[:200]}...") # Truncate for readability
    elif message.type == "ai":
        print(f"\n[Final Answer]: {message.content}")
```

<p>
<strong>Architectural Insight:</strong>
The <strong>ReAct</strong> pattern introduces meaningful <strong>autonomy</strong>.
Rather than hardcoding a verification procedure, the developer provides
<strong>capabilities</strong> (e.g., Search) and a clear <strong>intent</strong>
(e.g., verify a claim). The agent then determines <em>how</em> to reach the goal
through iterative reasoning and tool use.
</p>

<p>
This autonomy comes with a performance trade-off: ReAct agents can be
<strong>latency-intensive</strong> due to the sequential nature of
<strong>“Think → Call → Wait”</strong> loops.
</p>

<p>
In production, bounding the agent’s execution is essential. Setting a strict
<code>recursion_limit</code> (maximum steps) helps prevent runaway behavior and
infinite loops when the agent becomes uncertain or cycles on ambiguous evidence.
</p>


### 2.3 Custom Logic Pattern (Deterministic Routing)

![QuickCURD](https://docs.cloud.google.com/static/architecture/images/choose-design-pattern-agentic-ai-system-custom.svg)<br>
Credits: Google

<h3>Custom Logic Pattern</h3>

<p>
<strong>Definition:</strong>
Not all decisions in an AI system require an LLM. The <strong>Custom Logic</strong>
pattern embeds deterministic code (e.g., Python functions, if/else conditions, rule
checks) directly into the graph’s control flow. This pattern optimizes
<strong>cost</strong> and <strong>latency</strong> by offloading binary or rule-based
decisions to the <strong>CPU</strong> rather than invoking GPU-backed LLM reasoning.
</p>

<p>
<strong>Role in the TalentFlow Pipeline: Logic Filtering</strong><br/>
Before a candidate is advanced to a human interviewer, they must satisfy
non-negotiable legal or logistical constraints. Examples include:
<strong>“Must have a valid visa”</strong> or
<strong>“Must be within timezone ± 3 hours”</strong>.
</p>

<p>
Using an LLM to decide these constraints is both <strong>wasteful</strong> and
<strong>error-prone</strong>, since the model may hallucinate or treat hard constraints
as negotiable. A <strong>Custom Logic node</strong> enforces such rules
<strong>rigorously and deterministically</strong>.
</p>

<p>
<strong>Technical Implementation:</strong><br/>
A conditional routing function is defined to inspect the current state and return
the <strong>name of the next node</strong>. This allows the graph to route candidates
based on hard constraints without an LLM call.
</p>


```python
from typing import TypedDict, Literal
from langgraph.graph import StateGraph, START, END

# --- Step 1: Define State ---
class FilterState(TypedDict):
    candidate_name: str
    visa_status: str # e.g., "Citizen", "Visa", "None"
    experience_years: int
    decision: str

# --- Step 2: Define Nodes ---
# These nodes represent the outcome of the logic filter.
def fast_track_node(state: FilterState):
    return {"decision": "FAST TRACK: High experience, valid visa."}

def manual_review_node(state: FilterState):
    return {"decision": "MANUAL REVIEW: Visa sponsorship required."}

def reject_node(state: FilterState):
    return {"decision": "AUTO REJECT: Insufficient experience."}

# --- Step 3: Define Custom Logic Function ---
# This function contains the business rules. It is deterministic.
def route_candidate(state: FilterState) -> Literal["fast_track", "manual_review", "reject"]:
    years = state["experience_years"]
    visa = state["visa_status"]

    print(f"Routing candidate: {state['candidate_name']} (Exp: {years}, Visa: {visa})")

    if years < 2:
        return "reject"
    elif years >= 5 and visa == "Citizen":
        return "fast_track"
    else:
        return "manual_review"

# --- Step 4: Build the Graph ---
workflow = StateGraph(FilterState)
workflow.add_node("fast_track", fast_track_node)
workflow.add_node("manual_review", manual_review_node)
workflow.add_node("reject", reject_node)

# Add conditional edges based on the logic function
workflow.add_conditional_edges(
    START,
    route_candidate,
    {
        "fast_track": "fast_track",
        "manual_review": "manual_review",
        "reject": "reject"
    }
)
workflow.add_edge("fast_track", END)
workflow.add_edge("manual_review", END)
workflow.add_edge("reject", END)

app_logic = workflow.compile()
display(Image(app_logic.get_graph().draw_mermaid_png()))
# --- Tutorial Execution ---
candidates = [
    {"candidate_name": "Alice", "visa_status": "Citizen", "experience_years": 7},
    {"candidate_name": "Bob", "visa_status": "Visa", "experience_years": 6},
    {"candidate_name": "Charlie", "visa_status": "Citizen", "experience_years": 1},
    {"candidate_name": "Diana", "visa_status": "Visa", "experience_years": 4}
]

print("--- Running Logic Filter ---")
for c in candidates:
    result = app_logic.invoke(c)
    print(f"Outcome: {result['decision']}\n")
```

<p>
<strong>Architectural Insight:</strong>
Hybrid architectures that combine <strong>Probabilistic AI</strong> (LLMs) with
<strong>Deterministic Logic</strong> (code) are superior to purely LLM-driven systems.
In practice, this pattern functions as a <strong>guardrail</strong>: it ensures that
no amount of prompt injection, ambiguous phrasing, or model confusion can override
<strong>critical business rules</strong>.
</p>

<p>
For example, deterministic eligibility checks prevent outcomes such as advancing a
candidate who is <strong>legally ineligible to work</strong>, regardless of how
convincing the candidate’s narrative appears in free-form text.
</p>


## 3. Workflow Control Patterns

<p>
<strong>Workflow patterns</strong> dictate the <strong>sequence</strong> and
<strong>topology</strong> of task execution. They determine whether tasks run
<strong>one after another</strong>, <strong>simultaneously</strong>, or
<strong>repeatedly</strong>—and therefore directly influence system
<strong>latency</strong>, <strong>cost</strong>, and <strong>operational reliability</strong>.
</p>


### 3.1 Sequential Pattern


![QuickCURD](https://docs.cloud.google.com/static/architecture/images/choose-design-pattern-agentic-ai-system-sequential.svg)<br>
Credits: Google

<h3>Sequential Pattern</h3>

<p>
<strong>Definition:</strong>
The <strong>Sequential</strong> pattern is the backbone of most process automation.
It organizes agents or tools into a <strong>linear pipeline</strong> where the output
of one node becomes the mandatory input for the next. This creates a
<strong>“chain of reasoning” at the architectural level</strong>, ensuring that a
task is decomposed into manageable, dependent steps.
</p>

<p>
<strong>Role in the TalentFlow Pipeline: Job Description (JD) Generation</strong><br/>
Creating a high-quality Job Description is inherently multi-stage and order-dependent:
</p>

<ul>
  <li>
    <strong>Analyze:</strong>
    Extract technical requirements from the hiring manager’s rough notes.
  </li>
  <li>
    <strong>Draft:</strong>
    Produce the prose of the job description in clear, professional language.
  </li>
  <li>
    <strong>Format:</strong>
    Convert the final text into <strong>HTML</strong> for the careers page.
    This must happen last—formatting cannot occur before content exists.
  </li>
</ul>

<p>
<strong>Technical Implementation:</strong><br/>
In <strong>LangGraph</strong>, the Sequential pattern is implemented by connecting
nodes using simple, unconditional edges via <code>add_edge</code>, ensuring each step
executes strictly after the previous one.
</p>


```python
# --- Step 1: Define Sequence State ---
class JobDescriptionState(TypedDict):
    hiring_manager_notes: str
    requirements_analysis: str
    draft_text: str
    final_html: str

# --- Step 2: Define Sequential Nodes ---
def analyze_requirements(state: JobDescriptionState):
    print("--> Step 1: Analyzing Requirements")
    # Simulate LLM extraction
    analysis = llm.invoke(f"Extract key technical skills from: {state['hiring_manager_notes']}")
    return {"requirements_analysis": analysis.content}

def draft_content(state: JobDescriptionState):
    print("--> Step 2: Drafting Description")
    # Depends on previous step's output
    draft = llm.invoke(f"Write a 50-word engaging job description for: {state['requirements_analysis']}")
    return {"draft_text": draft.content}

def format_html(state: JobDescriptionState):
    print("--> Step 3: Formatting HTML")
    # Formatting only
    html = f"<div><h1>Job Opening</h1><p>{state['draft_text']}</p></div>"
    return {"final_html": html}

# --- Step 3: Build Linear Graph ---
seq_workflow = StateGraph(JobDescriptionState)
seq_workflow.add_node("analyze", analyze_requirements)
seq_workflow.add_node("draft", draft_content)
seq_workflow.add_node("format", format_html)

# Linear Edges: Start -> Analyze -> Draft -> Format -> End
seq_workflow.add_edge(START, "analyze")
seq_workflow.add_edge("analyze", "draft")
seq_workflow.add_edge("draft", "format")
seq_workflow.add_edge("format", END)

app_seq = seq_workflow.compile()
display(Image(app_seq.get_graph().draw_mermaid_png()))
# --- Tutorial Execution ---
input_notes = "We need a Python expert who knows LangGraph, AI agents, and can work remote."
result = app_seq.invoke({"hiring_manager_notes": input_notes})
print(f"\nFinal Output:\n{result['final_html']}")
```

We value remote work and seek a tech-savvy individual passionate about AI advancements. If you thrive in a dynamic environment and embrace challenges, we want to hear from you!</p></div>

<p>
<strong>Architectural Insight:</strong>
<strong>Sequential</strong> patterns maximize <strong>predictability</strong>.
In regulated domains such as HR, <strong>auditability</strong> is a primary requirement.
A sequential pipeline produces a clear, step-by-step trace of transformations, making
it straightforward to identify where a process degraded or failed.
</p>

<p>
For example, if the final <strong>HTML output</strong> is incorrect, the system can
inspect intermediate state (e.g., <code>draft_text</code>) to determine whether the
error originated during <strong>drafting</strong> or during <strong>formatting</strong>.
This traceability improves debugging, compliance reporting, and operational reliability.
</p>


### 3.2 Parallel Pattern (Map-Reduce / Fan-Out)

![QuickCURD](https://docs.cloud.google.com/static/architecture/images/choose-design-pattern-agentic-ai-system-parallel.svg)<br>
Credits: Google

<h3>Parallel Pattern</h3>

<p>
<strong>Definition:</strong>
The <strong>Parallel</strong> pattern enables multiple agents or tools to execute
<strong>simultaneously</strong>, significantly reducing total
<strong>wall-clock time</strong> for independent operations. In distributed systems,
this approach is commonly referred to as <strong>Fan-Out / Fan-In</strong> or
<strong>Map-Reduce</strong>.
</p>

<p>
In this pattern, a task is split into independent sub-tasks, executed concurrently,
and then <strong>aggregated (reduced)</strong> into a unified result. Parallelism
improves responsiveness but introduces coordination and aggregation complexity.
</p>

<p>
<strong>Role in the TalentFlow Pipeline: Candidate Sourcing</strong><br/>
When a new role opens, recruiters must search across multiple external platforms such
as <strong>LinkedIn</strong>, <strong>GitHub</strong>, <strong>StackOverflow</strong>,
and <strong>Twitter</strong>.
</p>

<p>
Executing these searches sequentially
(<em>search LinkedIn → wait → search GitHub → wait</em>) is inefficient and
unnecessarily slow. The <strong>Parallel</strong> pattern allows the
<strong>Sourcing Agent</strong> to spawn multiple search tasks at the same time and
aggregate all discovered candidates into a single consolidated list.
</p>

<p>
<strong>Technical Implementation:</strong><br/>
In <strong>LangGraph</strong>, parallel execution is supported via the
<strong>Send API</strong>. A <em>map node</em> emits multiple
<code>Send</code> objects, each dynamically dispatching work to a target node.
</p>

<p>
To merge results from parallel branches, the graph state defines a
<strong>reducer</strong> (for example, using <code>operator.add</code>) that combines
outputs—such as candidate lists—into a single state field during the fan-in phase.
</p>


```python
import operator
from typing import Annotated, List, TypedDict
from langgraph.types import Send
from langgraph.graph import StateGraph, START, END

# --- Step 1: Define State with Reducer ---
# 'found_candidates' uses operator.add to merge lists from parallel branches.
class SourcingState(TypedDict):
    platforms: List[str]
    found_candidates: Annotated[List[str], operator.add]

# --- Step 2: Define Worker Node ---
# This node runs in parallel instances.
def platform_scraper(state: dict):
    platform = state["platform"] # Passed via Send
    print(f"  : Scraping {platform}...")
    # Simulate finding a candidate specific to the platform
    return {"found_candidates": [f"Candidate from {platform}"]}

# --- Step 3: Define Map (Fan-Out) Logic ---
def distribute_searches(state: SourcingState):
    print(f"--> Orchestrator: Distributing work to {len(state['platforms'])} platforms.")
    # Return a list of Send objects.
    # Each Send targets 'scrape_node' with a specific payload.
    return [Send(node="scrape_node", arg={"platform": p}) for p in state["platforms"]]

# --- Step 4: Build Graph ---
workflow_par = StateGraph(SourcingState)
workflow_par.add_node("scrape_node", platform_scraper)

# Conditional edge from Start to the Map logic
workflow_par.add_conditional_edges(
    START,
    distribute_searches,
    {
        "scrape_node": "scrape_node"
    }
)
workflow_par.add_edge("scrape_node", END)

app_par = workflow_par.compile()
display(Image(app_par.get_graph().draw_mermaid_png()))
# --- Tutorial Execution ---
platforms = ["LinkedIn", "Indeed", "Glassdoor"]
print("--- Starting Parallel Sourcing ---")
result = app_par.invoke({"platforms": platforms, "found_candidates": []})
print(f"--- Aggregated Results ---")
print(result["found_candidates"])
```

<p>
<strong>Architectural Insight:</strong>
<strong>Parallelism</strong> is essential for scalability. The <strong>Send API</strong>
makes the graph <strong>dynamic</strong>, allowing it to adapt to the input size.
Whether the system searches <strong>3 platforms</strong> or <strong>30</strong>, the
same pattern can fan out automatically without redesigning the graph.
</p>

<p>
The primary trade-off is <strong>error-handling complexity</strong>. If one parallel
branch fails, the system must decide whether the entire sourcing job fails or whether
it proceeds with <strong>partial results</strong>.
</p>

<p>
In robust implementations, each branch should be isolated with its own
<strong>defensive handling</strong> (e.g., <code>try/except</code> and structured error
returns), so failures remain localized and the aggregator can make an explicit decision
about completeness vs. availability.
</p>


### 3.3 Loop Pattern


![QuickCURD](https://docs.cloud.google.com/static/architecture/images/choose-design-pattern-agentic-ai-system-loop.svg)<br>
Credits: Google

<h3>Loop Pattern</h3>

<p>
<strong>Definition:</strong>
The <strong>Loop</strong> pattern enables an agentic workflow to repeat a set of
actions until a specific <strong>state condition</strong> is satisfied. Unlike
<strong>ReAct</strong> (which loops on reasoning and tool use), the Loop pattern
typically iterates over <strong>data items</strong> (batch processing) or
<strong>quality conditions</strong> (retry logic).
</p>

<p>
Because looping introduces the possibility of runaway execution, it carries an
inherent risk of <strong>infinite loops</strong>. Robust implementations therefore
require strict, explicit <strong>termination conditions</strong>.
</p>

<p>
<strong>Role in the TalentFlow Pipeline: Initial Screening (Batch Processing)</strong><br/>
HR screening often occurs in batches of <strong>50–100 resumes</strong>. Spawning a
fresh end-to-end agent workflow for each resume can be inefficient and
resource-intensive. A <strong>Loop</strong> pattern allows a single
<strong>Screener Agent</strong> to iterate through a list of candidate IDs:
process one candidate, update state (e.g., remove from the queue, store results),
and repeat until the queue is empty.
</p>

<p>
<strong>Technical Implementation:</strong><br/>
In <strong>LangGraph</strong>, this is implemented via a <strong>conditional edge</strong>
that checks whether <code>pending_candidates</code> is empty. If candidates remain,
the graph routes back to the processing node; if not, it transitions to completion.
</p>


```python
class BatchProcessingState(TypedDict):
    pending_ids: List[str]
    processed_count: int
    results: Annotated[List[str], operator.add]

# --- Step 1: Processing Node ---
def screen_candidate(state: BatchProcessingState):
    # Get the next candidate
    current_id = state["pending_ids"][0]
    remaining_ids = state["pending_ids"][1:]

    print(f"Processing candidate ID: {current_id}")

    # Simulate screening logic
    result_str = f"ID {current_id}: Screened"

    # Update state: Remove processed ID, add result, increment count
    return {
        "pending_ids": remaining_ids,
        "processed_count": state["processed_count"] + 1,
        "results": [result_str]
    }

# --- Step 2: Loop Condition ---
def check_remaining(state: BatchProcessingState):
    if len(state["pending_ids"]) > 0:
        return "continue"
    return "stop"

# --- Step 3: Build Graph ---
workflow_loop = StateGraph(BatchProcessingState)
workflow_loop.add_node("screener", screen_candidate)

workflow_loop.add_edge(START, "screener")

# The Loop: From screener, check condition. If continue, go back to screener.
workflow_loop.add_conditional_edges(
    "screener",
    check_remaining,
    {
        "continue": "screener",
        "stop": END
    }
)

app_loop = workflow_loop.compile()
display(Image(app_loop.get_graph().draw_mermaid_png()))
# --- Tutorial Execution ---
batch = ["C001", "C002", "C003", "C004", "C005"]
print("--- Starting Batch Screening ---")
res = app_loop.invoke({"pending_ids": batch, "processed_count": 0, "results": []})
print(f"Total Processed: {res['processed_count']}")
```

<p>
<strong>Architectural Insight:</strong>
The <strong>Loop</strong> pattern introduces <strong>state persistence</strong>
challenges. As the loop iterates, the accumulated artifacts—such as message history,
intermediate outputs, or a growing results list—can expand continuously.
</p>

<p>
In high-volume loops (e.g., processing <strong>10,000+ items</strong>), the state object
may exceed practical limits, including <strong>in-memory constraints</strong> and
<strong>LLM context window</strong> capacity. This can degrade performance, increase
cost, and destabilize execution.
</p>

<p>
For large-scale batch workflows, a robust approach is to store operational state in an
<strong>external database</strong> (e.g., processed IDs, partial results, checkpoints),
while using the graph primarily for <strong>control-flow orchestration</strong> rather
than as the long-term storage container.
</p>


## 4. Quality & Refinement Patterns

<p>
These patterns focus on <strong>elevating output quality</strong> through
<strong>self-correction</strong> and <strong>multi-perspective evaluation</strong>,
allowing agentic systems to identify errors, challenge assumptions, and refine
results before they are consumed downstream.
</p>

### 4.1 Iterative Refinement Pattern

![QuickCURD](https://docs.cloud.google.com/static/architecture/images/choose-design-pattern-agentic-ai-system-iterative-refinement.svg)<br>
Credits: Google

<h3>Iterative Refinement Pattern</h3>

<p>
<strong>Definition:</strong>
<strong>Iterative Refinement</strong> mirrors the human drafting process. An agent
produces an initial output, critiques it against an explicit set of criteria, and then
uses that critique to generate an improved version. This loop continues until a
<strong>quality threshold</strong> is met or a <strong>maximum iteration count</strong>
is reached.
</p>

<p>
<strong>Role in the TalentFlow Pipeline: Profile Optimization</strong><br/>
When presenting a candidate to a hiring manager, the summary must be
<strong>concise</strong>, <strong>role-relevant</strong>, and emphasize differentiators.
A first-pass summary is often generic. An Iterative Refinement agent improves quality
through repeated drafting and evaluation, for example:
</p>

<ul>
  <li><strong>Draft</strong> an initial candidate summary.</li>
  <li><strong>Check</strong> it against a criterion such as: <em>“Does this highlight leadership?”</em></li>
  <li>If the criterion is not met, <strong>rewrite</strong> using the critique as guidance.</li>
</ul>

<p>
<strong>Technical Implementation:</strong><br/>
To prevent infinite loops and uncontrolled spending, the state tracks an
<code>iteration_count</code>. This creates a hard stop that forces termination even if
the quality check remains ambiguous—effectively addressing the practical implications
of the <strong>halting problem</strong> in iterative agent workflows.
</p>


```python
from IPython.display import Image, display
class RefinementState(TypedDict):
    original_text: str
    draft_summary: str
    critique: str
    iteration: int

def generate_draft(state: RefinementState):
    # First pass or subsequent passes
    if state["iteration"] == 0:
        msg = f"Summarize this resume profile: {state['original_text']}"
    else:
        msg = f"Improve this summary based on critique: {state['critique']}. Current Draft: {state['draft_summary']}"

    response = llm.invoke(msg)
    return {"draft_summary": response.content, "iteration": state["iteration"] + 1}

def critique_draft(state: RefinementState):
    print(f"--> Critiquing Draft (Iteration {state['iteration']})")
    # Simulate a critique - in production this would be another LLM call
    draft = state["draft_summary"]

    # Hard rule: Must mention 'Leadership'
    if "Leadership" not in draft and state["iteration"] < 3:
        return {"critique": "Missing mention of Leadership skills."}
    else:
        return {"critique": "SATISFIED"}

def quality_gate(state: RefinementState):
    if state["critique"] == "SATISFIED" or state["iteration"] >= 3:
        return END
    return "generate"

workflow_refine = StateGraph(RefinementState)
workflow_refine.add_node("generate", generate_draft)
workflow_refine.add_node("critique", critique_draft)

workflow_refine.add_edge(START, "generate")
workflow_refine.add_edge("generate", "critique")
workflow_refine.add_conditional_edges("critique", quality_gate)

app_refine = workflow_refine.compile()
display(Image(app_refine.get_graph().draw_mermaid_png()))
# --- Tutorial Execution ---
raw_profile = "Experienced developer. Good at coding. Managed a team of 5 people in 2020."
print("--- Starting Refinement Loop ---")
final = app_refine.invoke({"original_text": raw_profile, "iteration": 0})
print(f"Final Summary: {final['draft_summary']}")
```

<p>
<strong>Architectural Insight:</strong>
<strong>Iterative Refinement</strong> shifts the burden of quality from
<strong>prompt engineering</strong> to <strong>system architecture</strong>. Rather
than attempting to design a single, perfect “god prompt,” the system explicitly
assumes the first draft will be imperfect and embeds a structured mechanism to
improve it.
</p>

<p>
This approach aligns with <strong>System 2 thinking</strong> from cognitive science:
<strong>slow</strong>, <strong>deliberative</strong>, and reflective reasoning that
prioritizes accuracy and depth over immediacy.
</p>


### 4.2 Review and Critique Pattern (Scorer & Critic)

![QuickCURD](https://docs.cloud.google.com/static/architecture/images/choose-design-pattern-agentic-ai-system-review-critique.svg)<br>
Credits: Google

<h3>Review and Critique Pattern</h3>

<p>
<strong>Definition:</strong>
The <strong>Review and Critique</strong> pattern employs two explicitly distinct
personas: a <strong>Generator</strong> (or Scorer) and a <strong>Critic</strong>
(or Auditor). Unlike Iterative Refinement—where improvement often occurs within a
shared context—this pattern enforces a clear <strong>separation of duties</strong>.
</p>

<p>
This separation reduces <strong>bias confirmation</strong>, a failure mode in which
an agent implicitly defends its own output, and enables
<strong>adversarial evaluation</strong>, where assumptions, rationale, and decisions
are actively challenged.
</p>

<p>
<strong>Role in the TalentFlow Pipeline: Bias Auditing</strong><br/>
In candidate evaluation, one agent is responsible for
<strong>scoring</strong> a candidate based on skills and experience. A second,
independent <strong>DEI Auditor</strong> agent then reviews the scoring rationale to
ensure that no <strong>unconscious bias</strong> influenced the outcome.
</p>

<p>
Examples of bias signals include penalization based on
<strong>university prestige</strong>, <strong>employment gaps</strong>, or other
proxies unrelated to job performance. The Critic agent flags such issues and can
either annotate the score or trigger corrective action.
</p>


```python
class AuditState(TypedDict):
    candidate_profile: str
    initial_score: int
    score_rationale: str
    audit_flag: bool
    final_verdict: str

def scorer_agent(state: AuditState):
    print("--> Scorer: Evaluating Candidate")
    # Simulating a potentially biased or strict score
    return {
        "initial_score": 85,
        "score_rationale": "Great technical skills, but attended a non-target university."
    }

def auditor_agent(state: AuditState):
    print("--> Auditor: Checking for Bias")
    rationale = state["score_rationale"]

    # The auditor checks the rationale, not the code
    response = llm.invoke(
        f"Review this hiring rationale for bias: '{rationale}'. "
        "If biased, flag it. If fair, confirm it."
    )

    is_biased = "bias" in response.content.lower()
    return {"audit_flag": is_biased, "final_verdict": response.content}

workflow_audit = StateGraph(AuditState)
workflow_audit.add_node("scorer", scorer_agent)
workflow_audit.add_node("auditor", auditor_agent)

workflow_audit.add_edge(START, "scorer")
workflow_audit.add_edge("scorer", "auditor")
workflow_audit.add_edge("auditor", END)

app_audit = workflow_audit.compile()
display(Image(app_audit.get_graph().draw_mermaid_png()))

# --- Tutorial Execution ---
profile = "Technically strong candidate from State College."
res = app_audit.invoke({"candidate_profile": profile})
print(f"Initial Score: {res['initial_score']}")
print(f"Auditor Verdict: {res['final_verdict']}")
```

<p>
<strong>Architectural Insight:</strong>
The <strong>Review and Critique</strong> pattern embodies the principle of
<strong>Separation of Concerns</strong>. In high-compliance environments, introducing
a distinct <strong>Auditor</strong> step creates an explicit, inspectable record that
fairness and bias checks were performed as part of the decision process.
</p>

<p>
This separation also enables <strong>model stratification</strong>. A
cost-efficient model can be used for initial scoring, while a more
<strong>reasoning-intensive model</strong> (for example, GPT-4-class models) is
reserved for auditing and critique—optimizing for both
<strong>cost efficiency</strong> and <strong>decision quality</strong>.
</p>


## 5. Orchestration & Emergent Patterns

<p>
These <strong>advanced patterns</strong> govern <strong>complex interactions</strong>
across multiple agents, humans, and external sub-systems. They focus on
<strong>coordination</strong>, <strong>delegation</strong>, and
<strong>orchestration</strong>, enabling agentic systems to operate as
multi-actor environments rather than isolated decision-makers.
</p>


### 5.1 Coordinator Pattern (Router)

![QuickCURD](https://docs.cloud.google.com/static/architecture/images/choose-design-pattern-agentic-ai-system-coordinator.svg)<br>
Credits: Google

<h3>5.1 Coordinator Pattern (Router)</h3>

<p>
<strong>Definition:</strong>
The <strong>Coordinator</strong> functions as a central <strong>traffic controller</strong>
(or triage layer). It analyzes an incoming request and routes it to the appropriate
<strong>specialized worker</strong>. Unlike a Supervisor, the Coordinator typically
does not maintain long-term state or conduct extended back-and-forth; it focuses on
<strong>classification and handoff</strong>.
</p>

<p>
<strong>Role in the TalentFlow Pipeline: Process Routing</strong><br/>
HR operations receive large volumes of inbound messages and emails, such as:
<strong>“I want to apply”</strong>, <strong>“Where is my interview link?”</strong>,
and <strong>“What are the benefits?”</strong>.
</p>

<p>
A <strong>Coordinator</strong> identifies the intent and routes each request to the
correct downstream capability—for example:
<strong>Application Agent</strong>, <strong>Scheduling Agent</strong>, or a
<strong>Benefits RAG System</strong>.
</p>

<p>
<strong>Technical Implementation:</strong><br/>
This pattern is implemented using a <strong>router node</strong> that outputs the
<strong>destination node name</strong>. The graph then transitions to that node,
enabling clean separation between <em>triage</em> and <em>specialized execution</em>.
</p>


```python
def intent_classifier(state: dict):
    query = state["query"]
    print(f"Classifying Query: {query}")

    # Simple logic - in production use an LLM classifier
    if "interview" in query.lower():
        return "scheduling"
    elif "benefit" in query.lower():
        return "benefits"
    else:
        return "general_inbox"

def scheduling_node(state: dict):
    return {"response": "Routing to Scheduling System..."}

def benefits_node(state: dict):
    return {"response": "Retrieving 401k Information..."}

def inbox_node(state: dict):
    return {"response": "Ticket Created in General Inbox."}

workflow_router = StateGraph(dict)
workflow_router.add_node("scheduling", scheduling_node)
workflow_router.add_node("benefits", benefits_node)
workflow_router.add_node("general_inbox", inbox_node)

# The Conditional Edge is the Coordinator
workflow_router.add_conditional_edges(START, intent_classifier)
workflow_router.add_edge("scheduling", END)
workflow_router.add_edge("benefits", END)
workflow_router.add_edge("general_inbox", END)

app_router = workflow_router.compile()
display(Image(app_router.get_graph().draw_mermaid_png()))
# --- Tutorial Execution ---
q1 = "When is my interview?"
print(f"Query: {q1} -> {app_router.invoke({'query': q1})['response']}")
```

<p>
<strong>Architectural Insight:</strong>
The <strong>Coordinator</strong> pattern is fundamental to
<strong>decoupling</strong>. It cleanly separates the
<strong>user entry point</strong> from the <strong>backend implementation</strong>,
allowing the system to evolve without breaking the interface contract.
</p>

<p>
Because routing logic is centralized, new specialized agents
(for example, a <strong>“Visa Agent”</strong>) can be introduced without modifying
the user-facing workflow. As long as the classifier is updated to recognize the new
intent, the Coordinator can route requests appropriately—preserving extensibility,
maintainability, and long-term system stability.
</p>


### 5.2 Hierarchical Task Decomposition (Supervisor)

![QuickCURD](https://docs.cloud.google.com/static/architecture/images/choose-design-pattern-agentic-ai-system-hierarchical-task.svg)<br>
Credits: Google

<h3>Supervisor Pattern</h3>

<p>
<strong>Definition:</strong>
The <strong>Supervisor</strong> pattern implements a hierarchical structure in which a
central <strong>Manager</strong> agent oversees the state of a complex task and
delegates sub-tasks to specialized <strong>Worker</strong> agents. Unlike a pure
<strong>Router</strong>, the Supervisor maintains context and can reason across
multiple steps.
</p>

<p>
A Supervisor may call Worker A, evaluate the result, and then decide to call Worker B
based on the updated task state. Importantly, Workers typically report back to the
Supervisor—not directly to the user—so that orchestration remains centralized and
consistent.
</p>

<p>
<strong>Role in the TalentFlow Pipeline: Team Orchestration</strong><br/>
Interview coordination is inherently multi-step. A <strong>Recruiting Supervisor</strong>
must orchestrate multiple dependent actions, such as:
</p>

<ul>
  <li>
    Ask the <strong>Sourcing Agent</strong> to propose suitable candidates.
  </li>
  <li>
    Ask the <strong>Scheduling Agent</strong> to locate feasible time slots.
  </li>
  <li>
    Ask the <strong>Email Agent</strong> to send calendar invites and confirmations.
  </li>
</ul>

<p>
The Supervisor retains task memory—for example, <em>“Which candidate are we scheduling?”</em>—
across these steps, ensuring continuity and preventing state fragmentation across workers.
</p>

<p>
<strong>Technical Implementation:</strong><br/>
In <strong>LangGraph</strong>, the Supervisor is represented as a node that inspects
the current <strong>state</strong> and <strong>history</strong>, then selects the
<code>next_step</code> (i.e., the next worker or terminal action) via conditional routing.
</p>


```python
from langchain_core.messages import AIMessage, BaseMessage
from typing import List, TypedDict
from langgraph.graph import StateGraph, START, END

class TeamState(TypedDict):
    messages: List[BaseMessage]
    next_step: str

def supervisor_node(state: TeamState):
    msgs = state["messages"]
    last_msg = msgs[-1].content if msgs else ""

    # Logic: Source -> Schedule -> Finish
    if "Source a candidate" in last_msg and not any("Found" in msg.content for msg in msgs):
        return {"next_step": "sourcer"}
    elif any("Found" in msg.content for msg in msgs) and not any("Scheduled" in msg.content for msg in msgs):
        return {"next_step": "scheduler"}
    else:
        return {"next_step": "FINISH"}

def worker_sourcer(state: TeamState):
    print("  : Sourcing Candidate...")
    return {"messages": state["messages"] + [AIMessage(content="Found a candidate for the role.")]}

def worker_scheduler(state: TeamState):
    print("  : Scheduling Interview...")
    return {"messages": state["messages"] + [AIMessage(content="Scheduled an interview for the candidate.")]}

workflow_hier = StateGraph(TeamState)
workflow_hier.add_node("supervisor", supervisor_node)
workflow_hier.add_node("sourcer", worker_sourcer)
workflow_hier.add_node("scheduler", worker_scheduler)

workflow_hier.add_edge(START, "supervisor")

# Supervisor decides where to go
workflow_hier.add_conditional_edges(
    "supervisor",
    lambda x: x["next_step"],
    {
        "sourcer": "sourcer",
        "scheduler": "scheduler",
        "FINISH": END
    }
)

# Workers always report back to Supervisor
workflow_hier.add_edge("sourcer", "supervisor")
workflow_hier.add_edge("scheduler", "supervisor")

app_hier = workflow_hier.compile()
display(Image(app_hier.get_graph().draw_mermaid_png()))
# --- Tutorial Execution ---
print("--- Starting Hierarchical Flow ---")
final = app_hier.invoke({"messages": [HumanMessage(content="Source a candidate and schedule an interview.")], "next_step": ""})
print(f"Final State: {final}")
```

<p>
<strong>Architectural Insight:</strong>
<strong>Hierarchical</strong> patterns enable <strong>context encapsulation</strong>.
Worker agents (for example, sourcing or scheduling) remain
<strong>stateless and function-focused</strong>: they execute a well-defined task and
return results without retaining broader workflow memory.
</p>

<p>
The <strong>Supervisor</strong> absorbs the cognitive burden of maintaining
<strong>thread state</strong>, sequencing decisions, and interpreting intermediate
results. This mirrors real-world organizational hierarchies, where managers maintain
context while specialists execute tasks.
</p>

<p>
This separation promotes <strong>reusability</strong> of worker agents across
multiple workflows, simplifies testing, and reduces unintended coupling between
subsystems.
</p>


### 5.3 Swarm Pattern (Handoffs)

![QuickCURD](https://docs.cloud.google.com/static/architecture/images/choose-design-pattern-agentic-ai-system-swarm.svg)<br>
Credits: Google

<h3>Swarm Pattern</h3>

<p>
<strong>Definition:</strong>
In a <strong>Swarm</strong>, there is no central Supervisor. Agents operate
autonomously and transfer control to one another through <strong>peer-to-peer handoffs</strong>.
Agent A may recognize it cannot complete a task, identify that Agent B is better suited,
and then <strong>passes the baton</strong>—along with relevant state and intent—to Agent B.
</p>

<p>
This creates a <strong>decentralized</strong> topology in which collaboration emerges
from local decisions rather than centralized orchestration.
</p>

<p>
<strong>Role in the TalentFlow Pipeline: Hiring Committee</strong><br/>
A hiring committee naturally maps to swarm behavior. For example, a
<strong>Technical Interviewer</strong> might assess the candidate as technically strong,
but identify a concern around teamwork. Rather than escalating to a manager, the
Technical Interviewer directly hands off the case to the
<strong>Culture Interviewer</strong> with an explicit instruction such as:
<strong>“Probe on conflict resolution.”</strong>
</p>

<p>
This produces a more realistic committee dynamic: agents collaborate directly,
contributing specialized judgments while sharing context through explicit handoffs.
</p>

<p>
<strong>Technical Implementation:</strong><br/>
This pattern can be implemented using <strong>LangGraph’s</strong>
<code>Command</code> object, which enables explicit control-flow updates. A node can
return a <code>Command</code> that both (1) updates the shared state and (2) routes
execution to the next peer agent—making handoffs first-class and auditable.
</p>


```python
from langgraph.types import Command
from typing import TypedDict, List
from langgraph.graph import StateGraph, START, END

class SwarmState(TypedDict):
    candidate: str
    notes: str
    decision: str

def tech_agent(state: SwarmState):
    print("--- Tech Agent: Assessing Skills ---")
    # Tech check passes, but wants culture check
    return Command(
        goto="culture_agent",
        update={"notes": "Tech skills: 9/10. Please check soft skills."}
    )

def culture_agent(state: SwarmState):
    print("--- Culture Agent: Received Handoff ---")
    print(f"Incoming Notes: {state.get('notes')}")
    # Culture check finishes the process
    return Command(
        goto=END,
        update={"decision": "HIRE"}
    )

workflow_swarm = StateGraph(SwarmState)
workflow_swarm.add_node("tech_agent", tech_agent)
workflow_swarm.add_node("culture_agent", culture_agent)

# Entry point is Tech Agent
workflow_swarm.add_edge(START, "tech_agent")

app_swarm = workflow_swarm.compile()
display(Image(app_swarm.get_graph().draw_mermaid_png()))
# --- Tutorial Execution ---
print("--- Starting Swarm Hand-off ---")
# Initialize all state keys, even if empty
res = app_swarm.invoke({"candidate": "Jane", "notes": "", "decision": ""})
print(f"Final Decision: {res['decision']}")
```

<p>
<strong>Architectural Insight:</strong>
<strong>Swarm</strong> patterns can exhibit <strong>emergent behavior</strong>. They are
highly flexible and scalable because they remove a central bottleneck (a Supervisor)
and allow peers to collaborate dynamically based on local decisions.
</p>

<p>
The trade-off is governance risk. Without centralized control, the system can drift into
<strong>spaghetti routing</strong>, where agents repeatedly hand tasks back and forth and
fail to converge. This makes <strong>termination</strong>, <strong>accountability</strong>,
and <strong>debuggability</strong> harder than in hierarchical designs.
</p>

<p>
For Swarm architectures, strong <strong>observability</strong> is essential. Tracing and
visualization tools must capture the exact execution path—who handed off to whom, what
state was transferred, and why each decision was made—so the system remains auditable and
operationally controllable.
</p>


### 5.4 Human-in-the-Loop (HITL)

![QuickCURD](https://docs.cloud.google.com/static/architecture/images/choose-design-pattern-agentic-ai-system-human-in-the-loop.svg)<br>
Credits: Google

<h3>Human-in-the-Loop (HITL) Pattern</h3>

<p>
<strong>Definition:</strong>
<strong>Human-in-the-Loop (HITL)</strong> integrates human authorization directly into an
automated graph. The agent executes until a critical decision point, then
<strong>pauses</strong> execution, persists its state, and <strong>waits for human input</strong>.
Once input is provided, the graph <strong>resumes from the exact state</strong> where it stopped.
</p>

<p>
This pattern is essential when automation must be bounded by accountability and explicit
approval gates.
</p>

<p>
<strong>Role in the TalentFlow Pipeline: Offer Authorization</strong><br/>
An agent may determine that a candidate is hireable and generate a draft offer letter.
However, sending a financial contract is a <strong>high-risk, high-impact action</strong>.
The system must pause, present the draft to a <strong>Human HR Manager</strong>, and wait for a
clear decision such as <strong>“Approve”</strong> or <strong>“Revise”</strong> before invoking
any outbound email or dispatch tool.
</p>

<p>
<strong>Technical Implementation:</strong><br/>
In <strong>LangGraph</strong>, HITL is typically implemented using a
<strong>checkpointer</strong> to persist state and <code>interrupt_before</code> to halt the
graph before a sensitive node. After the human decision is received, the graph continues
execution using the stored checkpointed state.
</p>


```python
from langgraph.checkpoint.memory import InMemorySaver

# 1. Setup Memory for State Persistence
memory = InMemorySaver()

class OfferState(TypedDict):
    draft_amount: int
    manager_approval: str
    status: str

def draft_offer(state: OfferState):
    print("  : Drafting Offer for $120k...")
    return {"draft_amount": 120000}

def send_email(state: OfferState):
    # This node only runs AFTER human approves
    if state["manager_approval"] == "YES":
        return {"status": "SENT"}
    else:
        return {"status": "REJECTED"}

workflow_hitl = StateGraph(OfferState)
workflow_hitl.add_node("draft", draft_offer)
workflow_hitl.add_node("send", send_email)

workflow_hitl.add_edge(START, "draft")
workflow_hitl.add_edge("draft", "send")
workflow_hitl.add_edge("send", END)

# 2. Compile with Interrupt
# The graph will PAUSE before entering 'send' node.
app_hitl = workflow_hitl.compile(checkpointer=memory, interrupt_before=["send"])

# --- Tutorial Execution ---
thread_config = {"configurable": {"thread_id": "offer_101"}}

print("--- Phase 1: Run until Interrupt ---")
app_hitl.invoke({"manager_approval": "PENDING"}, config=thread_config)

print("\n: Waiting for Manager...")
#... Human reviews the $120k draft...

print("--- Phase 2: Human Updates State ---")
# Human acts via API (simulated here using update_state)
app_hitl.update_state(thread_config, {"manager_approval": "YES"})
display(Image(app_hitl.get_graph().draw_mermaid_png()))
print("--- Phase 3: Resume Execution ---")
# Invoke with None resumes from the paused state
final = app_hitl.invoke(None, config=thread_config)
print(f"Final Status: {final['status']}")
```

<p>
<strong>Architectural Insight:</strong>
<strong>Human-in-the-Loop (HITL)</strong> is the bridge between
<strong>automation</strong> and <strong>accountability</strong>. It transforms an
agentic system from a “black box” into a “glass box” by forcing critical decisions
to pass through an explicit human authorization gate, with state preserved and
reviewable.
</p>

<p>
In enterprise deployments, the pause must be durable. Instead of an
<code>InMemorySaver</code>, production systems typically use
<strong>persistent storage</strong> (e.g., <strong>Postgres</strong> or <strong>Redis</strong>)
so execution can remain halted for extended periods—days or weeks—while awaiting an
approver (for example, a manager who is on vacation). When the human responds, the
workflow resumes from the exact checkpointed state.
</p>


## 6. Conclusion: The Future of Enterprise Agency

<p>
The <strong>“TalentFlow”</strong> architecture demonstrates that there is no single
<strong>“best” agentic pattern</strong>. Successful enterprise implementations
require a <strong>polymorphic approach</strong>, where different patterns are applied
strategically based on task characteristics, risk, and cognitive demands.
</p>

<ul>
  <li>
    <strong>Single-Agent / Sequential</strong> patterns are best suited for
    <strong>compliance-heavy, deterministic tasks</strong>, such as
    Resume Parsing and Job Description Generation.
  </li>
  <li>
    <strong>ReAct / Parallel</strong> patterns excel in
    <strong>information gathering and research</strong>, including
    Candidate Sourcing and Claim Verification.
  </li>
  <li>
    <strong>Swarm / Hierarchical</strong> patterns are required for
    <strong>complex reasoning and multi-step orchestration</strong>, such as
    interview coordination and hiring committee decision-making.
  </li>
  <li>
    <strong>Custom Logic / Human-in-the-Loop (HITL)</strong> patterns function as
    <strong>safety valves and governance guardrails</strong>, particularly for
    eligibility filtering and final offer authorization.
  </li>
</ul>

<p>
As <strong>LLM reasoning costs continue to decline</strong> and
<strong>context windows expand</strong>, a gradual shift toward more
<strong>decentralized Swarm architectures</strong> is anticipated—where specialized
micro-agents collaborate dynamically to solve complex problems.
</p>

<p>
However, the <strong>Human-in-the-Loop</strong> pattern will remain the
<strong>ultimate safeguard</strong> in high-stakes domains such as HR. While AI
systems may execute the work, <strong>humans retain accountability and responsibility</strong>
for the outcomes.
</p>

<p>
By mastering these <strong>thirteen agentic patterns</strong>, organizations can
move beyond experimental chatbots and engineer
<strong>robust, scalable, and auditable cognitive workforces</strong>.
</p>

<h4>Tradeoff Lenses</h4>

<ul>
  <li>
    <strong>Latency-first view:</strong>
    <strong>Parallel</strong> and <strong>Coordinator</strong> patterns can improve

    calls and orchestration overhead.
  </li>
  <li>
    <strong>Cost-first view:</strong>
    <strong>Sequential</strong> is usually cheaper; <strong>loops</strong> and
    <strong>Hierarchical</strong> approaches tend to be more expensive due to repeated
    reasoning steps and multi-agent/subtask execution.
  </li>
  <li>
    <strong>Reliability-first view:</strong>
    <strong>Review / Critique</strong> and <strong>Human-in-the-Loop</strong> patterns
    increase safety and quality by adding validation, but they trade off speed and cost.
  </li>
</ul>
